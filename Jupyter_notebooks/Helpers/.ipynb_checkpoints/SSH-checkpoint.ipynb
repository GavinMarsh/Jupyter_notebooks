## SSH to Amazon Lightsail instance from Windows10

#### Download Putty
https://www.chiark.greenend.org.uk/~sgtatham/putty/latest.html


#### Step 1: Get the Private Key
Log into your Lightsail account.

On the top right, click on Account > Account.

Navigate to the SSH Keys tab.

Click the Download link which is next to the instance’s region.

(You can also upload or create new)

A LightsailDefaultKey-us-east-1.pem file will be downloaded.

#### Step 2: In PuTTYgen Generating an *.ppk File
Open the PuTTYgen client.

Go to File > Load private key.

In the Windows Browse dialog, switch from browsing just PuTTY Private Key Files (*.ppk) to All Files 

Open LightsailDefaultKey-us-east-1.pem file downloaded before.

Click on the Save private key button.

Click on Yes to Save it without the passphrase for convenience.

Give a file name and save it as a *.ppk file.

#### Step 3: SSHing with PuTTY
Launch PuTTY

Copy your website’s static IP address

Details of this can be found by clicking home>then the name of your Instance.

It will say Public IP:`//26.343.4.564.6//`

Username: `ubuntu` (keep a not of this for later)

In PuTTY, paste the IP in the Host Name field and `Save` so you can reload another day.

From the left, navigate to Connection > SSH > Auth.

Browse and open the *.ppk file that you created previously.

Click `Open`.

Confirm the trust connection dialog.

For Login as: enter the username (the username you kept from earlier).


# Install GUI onto Ubuntu18.94 server - Getting Started
https://comtechies.com/how-to-set-up-gui-on-amazon-ec2-ubuntu-server.html


#### Step 1: Connect to the server using ssh client putty. How To Connect Ubuntu Server Using Putty

#### Step 2: Make sure that RDP port is enabled on the Ubuntu instance in which you are going to install lxde. Update the server and install lxde using the following commands.

`sudo apt-get update -y`

`sudo apt-get install lxde -y`

#### Step 3:  Install xrdp to establish a remote desktop connection since you can’t have a GUI using putty. Use the following command to install xrdp

`sudo apt-get install xrdp -y`

#### Step 4: Set a password for the default user “ubuntu”, as remote desktop connection requires username and password.

`sudo passwd ubuntu`

Note: You can set up your own user and password for remote desktop

#### Step 5: Start the windows remote desktop client and enter the public DNS or the public IP of your server instance and hit connect.

Make sure you click the drop down menue and enter your user name: ubuntu

Save all the credentials for next time.


#### Step 6: Enter the username and password of the server instance that we created in step 5 and hit ok. Make sure RDP 3389 is enabled in your security group.

On lightsail the only way to do this is to select Application: `All protocals`

Note: If you are looking for more Linux command line hands on, use pluralsight 10 days free account to learn Linux Command Line. https://comtechies.com/recommends/pluralsight-linux-cli-course

Bingo!! you can start using your Ubuntu server with GUI via the remote desktop connection.
